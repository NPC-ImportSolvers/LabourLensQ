```python
import pytesseract
from pdf2image import convert_from_path
import pdfplumber
import pandas as pd

# If using Windows, uncomment and set your Tesseract path
# pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

def extract_table_with_ocr(pdf_path, title_keyword):
    """
    Uses OCR to extract text from each PDF page, finds the target page, and extracts tables.
    
    :param pdf_path: Path to the PDF file.
    :param title_keyword: Keyword to locate the desired page.
    :return: DataFrame with extracted table data.
    """
    # Step 1: Convert PDF to images
    images = convert_from_path(pdf_path, poppler_path=r"C:\poppler-24.08.0\Library\bin")

    # Step 2: OCR each page to find the one with the title
    target_page = None
    for i, image in enumerate(images):
        text = pytesseract.image_to_string(image)  # Perform OCR
        print(f"================{i}================")
        print(text)
        if title_keyword.lower().replace(" ", "") in text.lower().replace("\n", " ").replace(" ", ""):
            target_page = i
            break

    if target_page is None:
        print(f"Title '{title_keyword}' not found in the document.")
        return None

    print(f"Title found on page {target_page + 1}")

    # Step 3: Extract table from the identified page
    with pdfplumber.open(pdf_path) as pdf:
        tables = pdf.pages[target_page].extract_tables()

    if not tables:
        print(f"No tables found on page {target_page + 1}.")
        return None

    # Step 4: Convert the first table found into a DataFrame
    df = pd.DataFrame(tables[0])

    return df

# Example Usage
pdf_path = "../Annual_Bulletin_Labour_force_AE/Annual_Bulletin_Labour_force_2017_AE.pdf"
title_keyword = "ECONOMICALLY ACTIVE POPULATION (15 YEARS & ABOVE) BY NATIONALITY, GENDER, & ECONOMIC ACTIVITY" #87
df = extract_table_with_ocr(pdf_path, title_keyword)

if df is not None:
    print(df)
```

```python
import pdfplumber
import os
from PyPDF2 import PdfReader, PdfWriter

def is_page_match(page_text, title_keyword, threshold=90):
    """Check if at least `threshold%` of words from the title are in the page text."""
    title_words = set(title_keyword.lower().split())
    page_words = set(page_text.lower().split())
    
    common_words = title_words.intersection(page_words)
    match_percentage = (len(common_words) / len(title_words)) * 100
    
    return match_percentage >= threshold, match_percentage

def extract_matching_pages(directory, title_keyword, output_pdf):
    """Scan PDFs in a directory, find matching pages, and compile them into one PDF."""
    pdf_writer = PdfWriter()
    
    # Iterate through all PDFs in the directory
    for file_name in os.listdir(directory):
        if file_name.endswith(".pdf"):
            pdf_path = os.path.join(directory, file_name)
            print(f"🔍 Searching in {file_name}...")

            with pdfplumber.open(pdf_path) as pdf:
                pdf_reader = PdfReader(pdf_path)  # Read the full PDF to extract raw pages
                for page_num, page in enumerate(pdf.pages, start=0):  # pdfplumber uses 0-based indexing
                    text = page.extract_text()
                    if not text:
                        continue  # Skip empty pages
                    
                    match, match_percentage = is_page_match(text, title_keyword)
                    print(f"📄 Page {page_num + 1}: Match {match_percentage:.2f}%")

                    if match:
                        print(f"✅ Adding Page {page_num + 1} from {file_name} ({match_percentage:.2f}% match)")
                        pdf_writer.add_page(pdf_reader.pages[page_num])  # Correct way to add a page

    # Save the compiled PDF
    if len(pdf_writer.pages) > 0:
        output_path = os.path.join(directory, f"{title_keyword}.pdf")
        with open(output_path, "wb") as out_file:
            pdf_writer.write(out_file)
        print(f"✅ Merged PDF saved as: {output_path}")
    else:
        print("⚠️ No matching pages found in any PDF.")

# Example Usage
directory = "../Annual_Bulletin_Labour_force_AE/"
title_keyword = "WORKERS IN PAID EMPLOYMENT (15 YEARS & ABOVE) AND MONTHLY AVERAGE WAGE (Q.R.), BY GENDER & ECONOMIC ACTIVITY Agriculture Mining Construction Information Wholesale Education"
output_pdf = f"{title_keyword}.pdf"

extract_matching_pages(directory, title_keyword, output_pdf)
```

```python
import pdfplumber
import pandas as pd
import re

def extract_text_from_pdf(pdf_path, page_number):
    """ Extracts text from a specific page in a PDF """
    with pdfplumber.open(pdf_path) as pdf:
        if page_number > len(pdf.pages):
            raise ValueError(f"Page {page_number} does not exist in the document.")
        
        page = pdf.pages[page_number]  # Convert to zero-based index
        text = page.extract_text()
        print(text)
        if not text:
            raise ValueError("Text extraction failed. The page might contain only images or scanned text.")
        
        return text

def clean_and_split_text(text):
    """ Cleans text and splits into lines """
    lines = text.split("\n")
    return [line.strip() for line in lines if line.strip()]

def extract_table_data(lines):
    """ Extracts structured table data from lines using regex """
    extracted_data = []
    
    for line in lines:
        # Match column names and numbers (handling commas in numbers)
        matches = re.findall(r"([\w\s]+)\s+([\d,]+(?:\.\d+)?)", line)
        for match in matches:
            column_name = match[0].strip()
            value = match[1].replace(",", "")  # Remove commas from numbers
            extracted_data.append((column_name, value))
    
    return extracted_data

def create_dataframe(extracted_data):
    """ Converts extracted data into a Pandas DataFrame """
    if not extracted_data:
        print("No data extracted.")
        return None

    data_dict = {col: [val] for col, val in extracted_data}
    df = pd.DataFrame.from_dict(data_dict)
    return df

def extract_table_from_pdf(pdf_path, page_number):
    """ Main function to extract a table from a specific PDF page """
    raw_text = extract_text_from_pdf(pdf_path, page_number)
    lines = clean_and_split_text(raw_text)
    extracted_data = extract_table_data(lines)
    df = create_dataframe(extracted_data)
    return df

# Provide the PDF path and the page number
page_number = 92  # Change this to the actual page with the table

# Extract table from the PDF
df = extract_table_from_pdf(pdf_path, page_number)
df
```
